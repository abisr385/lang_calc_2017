## Exercise 3

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = [1.5]
```

```python
x = [1.5]
for i in range (10):
    next = (x[i] + 2/x[i])/2
    x.append(next)
```

```python
x
```

```python
plt.plot(x)
plt.title('Sequence x')
plt.xlabel('x')
plt.ylabel('y')
```

```python
z = [10]
for i in range (10):
    next = (z[i] + 2/z[i])/2
    z.append(next)
```

```python
z
```

```python
plt.plot(z)
plt.title('Sequence z')
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.plot(x)
plt.plot(z)
plt.title('graphs of x versus z')
plt.xlabel('x')
plt.ylabel('y')
```

x = 1.5 is a much closer approximation to solving the square root of 2 than z = 10. For that reason, the graph for x is more linear than that of z; with z = 10, we became close to solving the square root of 2 after 5 tries whereas with x = 1.5, we were close with the second try.
